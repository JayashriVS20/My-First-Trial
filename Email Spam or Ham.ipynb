```python
#Here category data is not numeric so there is need to converet it.
EmailDataset['Spam'] = EmailDataset['Category'].apply(lambda x: 1 if x=='spam' else 0)

EmailDataset
```

```python
#ther is no need of Category column 
EmailDataset1 = EmailDataset.drop('Category',axis = 'columns')
EmailDataset1
```

```python
#Now dividing row sample into traing and testing 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(EmailDataset1.Message,EmailDataset1.Spam,test_size = 0.2,random_state = 20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from tkinter import Tk, filedialog
from termcolor import colored

clf =Pipeline([
    ('v',CountVectorizer()),
    ('nb',MultinomialNB())
])

clf.fit(X_train,y_train)
"""sample_emails = [
     'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]

# Get user input for emails
user_emails = []
email_sub = int(input("Enter the number of emails you want to classify: "))

for i in range(email_sub):
    email = input(f"Enter email {i + 1}: ")
    user_emails.append(email)
    
"""

# Get user input for CSV file path
csv_file_path = input("Enter the path to the CSV file: ")

# Load the dataset from the user-provided CSV file
data = pd.read_csv(csv_file_path)

data.head()


# Predict using the trained classifier
sample_predictions = clf.predict(data['Message'])
spam_count = 0
ham_count = 0

for email, prediction in zip(data['Message'], sample_predictions):
    label = 'Spam' if prediction == 1 else 'Ham'
    color = 'red' if prediction == 1 else 'green'
    colored_label = colored(label, color)
    print(f"Email: {email}\nPrediction: {colored_label}\n")
        
        
    #print(f"Email: {email}\nPrediction: {label}\n")
    if label == 'Spam':
        spam_count += 1
    else:
        ham_count += 1

print("Total Spam:", spam_count)
print("Total Ham:", ham_count)


# Evaluate the model on the test set
y_pred = clf.predict(X_test)

score = clf.score(X_test,y_test)
print(f"Score is: {score}\n")





classification_rep = classification_report(y_test, y_pred)

print("Classification Report:\n", classification_rep)

#D:\Jayashri AVS\IT\Datasets for DS\spam.csv
```
